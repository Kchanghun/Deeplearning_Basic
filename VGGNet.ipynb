# VGG Network

## VGG16

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout

model_VGG16 = Sequential()

# Block 1
# 224,224,3
model_VGG16.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same', input_shape=(224,224,3)))
# 224,224,64
model_VGG16.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 224,224,64
model_VGG16.add(MaxPool2D((2,2), strides=(2,2)))
# 112,112,64

# Block 2
# 112,112,64
model_VGG16.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 112,112,128
model_VGG16.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 112,112,128
model_VGG16.add(MaxPool2D((2,2), strides=(2,2)))
# 56,56,128

# Block 3
# 56,56,128
model_VGG16.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 56,56,256
model_VGG16.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 56,56,256
model_VGG16.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 56,56,256
model_VGG16.add(MaxPool2D((2,2), strides=(2,2)))
# 28,28,256

# Block 4
# 28,28,256
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 28,28,512
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 28,28,512
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 28,28,512
model_VGG16.add(MaxPool2D((2,2), strides=(2,2)))
# 14,14,512

# Block 5
# 14,14,512
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 14,14,512
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 14,14,512
model_VGG16.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
# 14,14,512
model_VGG16.add(MaxPool2D((2,2), strides=(2,2)))
# 7,7,512

# Block 6
# 7,7,512
model_VGG16.add(Flatten())
# 25088
model_VGG16.add(Dense(4096, activation='relu'))
# 4096
model_VGG16.add(Dropout(0.5))
# 4096
model_VGG16.add(Dense(4096, activation='relu'))
# 4096
model_VGG16.add(Dropout(0.5))
# 4096
model_VGG16.add(Dense(1000, activation='softmax'))
# 1000

model_VGG16.summary()
```

# VGG19

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout

model_VGG19 = Sequential()

# Block 1
model_VGG19.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same', input_shape=(224,224,3)))
model_VGG19.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(MaxPool2D((2,2), strides=(2,2)))

# Block 2
model_VGG19.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(MaxPool2D((2,2), strides=(2,2)))

# Block 3
model_VGG19.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=256, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(MaxPool2D((2,2), strides=(2,2)))

# Block 4
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(MaxPool2D((2,2), strides=(2,2)))

# Block 5
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(Conv2D(filters=512, kernel_size=(3,3), strides=(1,1),\
                    activation='relu', padding='same'))
model_VGG19.add(MaxPool2D((2,2), strides=(2,2)))

# Block 6
model_VGG19.add(Flatten())
model_VGG19.add(Dense(4096, activation='relu'))
model_VGG19.add(Dropout(0.5))
model_VGG19.add(Dense(4096, activation='relu'))
model_VGG19.add(Dropout(0.5))
model_VGG19.add(Dense(1000, activation='softmax'))

model_VGG19.summary()
```
